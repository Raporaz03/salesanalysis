```python
#to import libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns
```

```python
#import csv file 
df=pd.read_csv('SalesData.csv',encoding='unicode_escape')
```

```python
df.head
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info
```

```python
#to check for null values 
pd.isnull(df)
```

```python
#to find number of null values in each column 
pd.isnull(df).sum()
```

```python
#to delete rows from null values 
df.shape
```

```python
df.dropna(inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#to change data type of Amount column from 
#float64 to int32

df['Amount']=df['Amount'].astype('int')
```

```python
#analysis on gender 
```

```python
ax= sns.countplot(x="Gender",data=df)
```

```python
ax=sns.countplot(x="Gender",data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen
```

```python
sns.set(rc={'figure.figsize':(7,5)})

sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data=sales_gen)

plt.show()
```

```python
ax=sns.countplot(x="Age Group",data =df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(x='Age Group', data =df , hue="Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount vs ae geoup 
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x="Age Group",y="Amount",data=sales_age)

plt.show()
```

```python
#analysis on state

sns.set(rc={'figure.figsize':(7,5)})

#total number of orders from top 10 sets 
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)

sns.barplot(x="State",y="Orders",data=sales_state)

plt.show()
```

```python
#total number of orders from top 5 sets 
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(5)

sns.barplot(x="State",y="Orders",data=sales_state)

plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,5)})

#total amount/sales from top 10 states
sales_state = df.groupby(["State"],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)

sns.barplot(x='State', y='Amount', data=sales_state)

plt.show()
```

```python
sales_state = df.groupby(["State"],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(5)

sns.barplot(x='State', y='Amount', data=sales_state)

plt.show()
```

```python
sns.set(rc={'figure.figsize':(7,5)})

ax=sns.countplot(x='Marital_Status',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ms =df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)

sns.barplot(x='Marital_Status',y='Amount',data=sales_ms,hue='Gender')

plt.show()
```

```python
#analysis on occupation 
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Occupation',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ocp =df.groupby(['Occupation'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(x='Occupation',y='Amount',data=sales_ocp)

plt.show()
```

```python
sales_ocp =df.groupby(['Occupation'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(7)

sns.set(rc={'figure.figsize':(7,5)})

sns.barplot(x='Occupation',y='Amount',data=sales_ocp)

plt.show()
```

```python
#analysis on product category 
```

```python
#top 10 most sold products 

sns.set(rc={'figure.figsize':(20,5)})

sales_pd = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.barplot(x='Product_ID',y="Orders",data=sales_pd)

plt.show()
```
